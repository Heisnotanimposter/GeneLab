```python
import numpy as np
import matplotlib.pyplot as plt
import random


# 평균 발생 횟수
lambda_ = 5

# 포아송 분포를 사용하여 시뮬레이션된 이벤트 발생 횟수
poisson_dist = np.random.poisson(lambda_, 1000)

# 결과 플롯
plt.hist(poisson_dist, bins=30, density=True)
plt.title('Poisson Distribution (λ = 5)')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()
```

```python

# 랜덤워크의 단계 수
steps = 1000

# 랜덤워크 위치 초기화
position = 0
walk = [position]

# 랜덤워크 시뮬레이션
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)

# 결과 플롯
plt.plot(walk)
plt.title('1D Random Walk')
plt.xlabel('Number of Steps')
plt.ylabel('Position')
plt.show()
```

```python


# 파라미터 설정
x = np.linspace(0, 10, 1000)  # x 축 범위
frequency = 1.0  # 주파수

# 사인과 코사인 파동
sin_wave = np.sin(2 * np.pi * frequency * x)
cos_wave = np.cos(2 * np.pi * frequency * x)

# 파동 합
combined_wave = sin_wave + cos_wave

# 결과 플롯
plt.plot(x, sin_wave, label='sin(2πfx)')
plt.plot(x, cos_wave, label='cos(2πfx)')
plt.plot(x, combined_wave, label='Combined Wave', color='black')
plt.title('Combination of Sine and Cosine Waves')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend()
plt.show()

```
